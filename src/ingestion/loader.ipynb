```python
import os
import sys
from dotenv import load_dotenv
import polars as pl
```

```python
load_dotenv()
HF_DATASET_PATH = os.getenv("HF-dataset")
```

```python
metadata = pl.read_parquet(f"hf://datasets/{HF_DATASET_PATH}/data/metadata.parquet")
```

```python
relationships = pl.read_parquet(f"hf://datasets/{HF_DATASET_PATH}/data/relationships.parquet")
```

```python
content = pl.read_parquet(f"hf://datasets/{HF_DATASET_PATH}/data/content.parquet")
```

```python
datas = {"metadata":metadata, "relationship":relationships, "content":content}
cols = {"metadata": datas["metadata"].columns,
       "relationship": datas["relationship"].columns,
         "content": datas["content"].columns
         }
max_len = max(len(df) for df in cols.values())
print(max_len)
for col in cols.values():
    col.extend([" "] * (max_len - len(col)))


table_columns = pl.DataFrame(cols)

print(table_columns)
```
